```python
prev = datetime.datetime.now()
info_df['num_subscribers'] = info_df['num_subscribers'].astype('int64')
info_df['num_reviews'] = info_df['num_reviews'].astype('int64')
info_df['num_comments'] = info_df['num_comments'].astype('int64')
info_df['num_lectures'] = info_df['num_lectures'].astype('int64')
info_df['num_reviews'] = info_df['num_reviews'].astype('int64')
info_df['last_update_date'] = ps.to_datetime(info_df['last_update_date'],errors='coerce')
now = datetime.datetime.now()
print(now - prev)
```

```python
def get_instructor_username(instructor_url):
    return str(instructor_url)[6:-2]
```

```python
info_df['instructor_url'] = info_df['instructor_url'].map(get_instructor_username)
```

```python
info_df.head()
```

```python
def get_curse_name(course_url):
    return str(course_url)[8:-1]
```

```python
info_df['course_url'] = info_df['course_url'].map(get_curse_name)
```

```python
info_df.rename(columns={'instructor_url':'instructor_username'},inplace=True)
```
